```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('http://meusite.mackenzie.br/rogerio/TIC/Projects/glassdoordata.csv')


plt.hist(data[data.gender == 'Male']['income'], label = 'Homens')
plt.hist(data[data.gender == 'Female']['income'], label = 'Mulheres')

plt.xlabel('Valor das rendas')
plt.ylabel('frequencia rendas')
plt.title('Distribuicao das rendas', fontsize = 14, weight = 'bold')
plt.legend(title = '$\\bf{Rendas}$')
plt.show()
```
